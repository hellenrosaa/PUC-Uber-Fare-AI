## Preparando os dados considerando 99 e Uber

product.csv

```python
import pandas as pd
```

```python
product_df = pd.read_csv('../data/product.csv', sep=';', names=['ProductID', 'ProviderID', 'CategoryID', 'Description'])
product_df = product_df.iloc[1:].reset_index(drop=True)

display(product_df)
```

```python
for col in product_df.columns:
    print(f"{col}: {product_df[col].dtype}")
```

```python
product_df['ProviderID'] = pd.to_numeric(product_df['ProviderID'], errors='coerce')
product_df['CategoryID'] = pd.to_numeric(product_df['CategoryID'], errors='coerce')
```

```python
print("------- Uber -------- ProviderID = 2 -------")
df_2 = product_df[product_df['ProviderID'] == 2]

print("ProductID únicos:", df_2['ProductID'].unique())
print("\nCategoryID únicos:", df_2['CategoryID'].unique())
print("\nDescription únicos:", df_2['Description'].unique())
```

```python
print("------- 99 -------- ProviderID = 3 -------")
df_3 = product_df[product_df['ProviderID'] == 3]

print("ProductID únicos:", df_3['ProductID'].unique())
print("\nCategoryID únicos:", df_3['CategoryID'].unique())
print("\nDescription únicos:", df_3['Description'].unique())
```
